```python
import numpy as np
import matplotlib.pyplot as plt
```

## First sample plot

The "Hello, world" of plotting: a sine curve

```python
x = np.arange(0, 2*np.pi, 0.01)
y = np.sin(x)
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('f(x) = sin(x)')
plt.grid()
```

```python
#Plot from Marion - oscillation

t = np.arange(0, 10, 0.1) #time interval
A = 0.5 #amplitude
omega = np.pi #frequency
phi = 0 #phase
y = A*np.sin(omega*t + phi) #oscillation

#Plot the oscillation
plt.plot(t, y)
plt.xlabel('t')
plt.ylabel('y(t)')
plt.grid()
plt.show()
```
