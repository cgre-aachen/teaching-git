```python
import numpy as np
import matplotlib.pyplot as plt
```

## First sample plot

The "Hello, world" of plotting: a sine curve

```python
x = np.arange(0, 2*np.pi, 0.01)
y = np.sin(x)
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('f(x) = sin(x)')
plt.grid()
```
